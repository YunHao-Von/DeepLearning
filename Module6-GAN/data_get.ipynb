```python
import os
import cv2
import sys
import numpy as np
```

```python
path = "samoyed_imgs/"
path2 = "dataset/"
images = os.listdir(path)
for j in range(len(os.listdir(path))):
    i = images[j]
    print(os.path.splitext(i))  # ('34474006827920603', '.png')
    if os.path.splitext(i)[1] == ".jpeg":
        img = cv2.imread(path + i)
        # print(img)
        new_imagename = str(j)+'.jpg'
        cv2.imwrite(path2 + new_imagename, img)

    elif os.path.splitext(i)[1] == ".png":
        img = cv2.imread(path + i)
        # print(img)
        new_imagename = str(j)+'.jpg'
        cv2.imwrite(path2 + new_imagename, img)

    elif os.path.splitext(i)[1] == ".JPG":
        img = cv2.imread(path + i)
        # print(img)
        new_imagename = str(j)+'.jpg'
        cv2.imwrite(path2 + new_imagename, img)

    elif os.path.splitext(i)[1] == ".PNG":
        img = cv2.imread(path + i)
        # print(img)
        new_imagename = str(j)+'.jpg'
        cv2.imwrite(path2 + new_imagename, img)

    elif os.path.splitext(i)[1] == ".jpg":
        img = cv2.imread(path + i)
        # print(img)
        cv2.imwrite(path2 + str(j)+'.jpg', img)
```
